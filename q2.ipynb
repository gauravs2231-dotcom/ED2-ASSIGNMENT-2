```python
import numpy as np
import matplotlib.pyplot as plt

# Normalising constansts to unity 
c=1
R=1
C=1
dotbz=0.3   # case a
dotbx=0.9   # case b
betas= np.array([0.1, 0.9, 0.999])
n0= 1 / np.sqrt(3)
t= np.linspace(0, 10, 500)
tret=t - R / c
active = np.where((tret >= 2) & (tret <= 7), 1.0, 0.0)  # like a pulse on/off


plt.figure()
for i in betas:
    d=(1 - n0*i)**3
    ex=C* dotbz*(1/3 - n0*i)/d *active
    plt.plot(t, ex, label='beta='+str(i))
plt.title('Case A (transverse): Ex vs time')
plt.xlabel('time')
plt.ylabel('Ex (arb units)')
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i in betas:
    d=(1 - n0*i)**3
    by=C*dotbz*n0*(1 - 2*n0*i)/d *active
    plt.plot(t, by,label='beta='+str(i))
plt.title('Case A (transverse): By vs time')
plt.xlabel('time')
plt.ylabel('By (arb units)')
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i in betas:
    d=(1 - n0*i)**3
    ex= -C*dotbx *(2/3)/d *active
    plt.plot(t, ex,label='beta='+str(i))
plt.title('Case B (longitudinal): Ex vs time')
plt.xlabel('time')
plt.ylabel('Ex (arb units)')
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
for i in betas:
    d=(1 - n0*i)**3
    by= -C * dotbx * n0/d * active
    plt.plot(t, by, label='beta='+str(i))
plt.title('Case B (longitudinal): By vs time')
plt.xlabel('time')
plt.ylabel('By (arb units)')
plt.grid(True)
plt.legend()
plt.show()



```
