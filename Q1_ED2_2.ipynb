```python
import numpy as np
import matplotlib.pyplot as plt

# Normalising constansts to unity 
c=q=b=k=1  
betas=[0.1, 0.9, 0.999] 

#  x-axis (For arbitrary Range) 
x=np.linspace(-5, 5, 400)

plt.figure()
for i in betas:
    gamma=1 / np.sqrt(1 - i**2)
    rprime=np.sqrt(b**2 + (gamma * x)**2)
    Ex=-k * gamma * x / (rprime**3)
    Ex=Ex / (np.max(np.abs(Ex)))      # normalizing to unity 
    plt.plot(x, Ex, label=f'β={i}')

plt.title("Ex vs Distance ")
plt.xlabel("x (distance along motion)")
plt.ylabel("Normalized Ex")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
for i in betas:
    gamma=1 / np.sqrt(1 - i**2)
    rprime=np.sqrt(b**2 + (gamma * x)**2)
    Ey=k * gamma * b / (rprime**3)
    Ey=Ey / np.max(np.abs(Ey))
    plt.plot(x, Ey, label=f'β={i}')

plt.title("Ey vs Distance")
plt.xlabel("x (distance along motion)")
plt.ylabel("Normalized Ey")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
for i in betas:
    gamma=1 / np.sqrt(1 - i**2)
    rprime=np.sqrt(b**2 + (gamma * x)**2)
    Ey=k * gamma * b / (rprime**3)
    Bz=beta * Ey  # Magnetic field
    Bz=Bz / np.max(np.abs(Bz))
    plt.plot(x, Bz, label=f'β={i}')

plt.title("Magnetic Field Bz vs distance x")
plt.xlabel("x (distance along motion)")
plt.ylabel("Normalized Bz")
plt.legend()
plt.grid(True)
plt.show()


```
