```python
import numpy as np
import matplotlib.pyplot as plt

vo= 1
alpha=1/100
w= 30
t= np.linspace(0, 1000, 2000)
f= vo * np.exp(alpha*t) * (2*alpha*w*np.cos(w*t) + (alpha**2 - w**2)*np.sin(w*t))
plt.plot(t, f)
plt.title('Radiation reaction force (normalized units)')
plt.xlabel('Time')
plt.ylabel('Radiation reaction force')
plt.show()
```
